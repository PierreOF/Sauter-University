```python
import pandas as pd
from pandasql import sqldf
```

```python
df_dados_covid = pd.read_csv('/caso_full.csv')
```

```python
df_dados_covid.shape
```

```python
traducoes = {
    "city": "Nome da cidade",
    "city_ibge_code": "codigo ibge da cidade",
    "date": "data",
    "epidemiological_week": "semana epidemiologica",
    "estimated_population": "populacao estimada",
    "estimated_population_2019": "populacao estimada_2019",
    "is_last": "e ultima",
    "is_repeated": "e repetida",
    "last_available_confirmed": "ultimos confirmados",
    "last_available_confirmed_per_100k_inhabitants": "ultimos confirmados por 100mil habitantes",
    "last_available_date": "ultima data_disponivel",
    "last_available_death_rate": "taxa_de_mortalidade_ultima",
    "last_available_deaths": "ultimas_mortes_disponiveis",
    "order_for_place": "ordem_para_o_local",
    "place_type": "tipo_de_local",
    "state": "estado",
    "new_confirmed": "novos_confirmados",
    "new_deaths": "novas_mortes"
}
```

```python

df_dados_covid = df_dados_covid.rename(columns=traducoes)
```

```python
df_dados_covid.sort_values(by='data', inplace=True)
df_dados_covid[['data']].tail(1)
```

```python
df_dados_covid.groupby('estado')[['ultimas_mortes_disponiveis']].mean().round(2)
```

```python
mortalidade_por_cidade = df_dados_covid.loc[df_dados_covid.groupby('estado')['taxa_de_mortalidade_ultima'].idxmax()][['Nome da cidade']]
mortalidade_por_cidade
```

```python
semana_especifica = 202103
df_semana_especifica = df_dados_covid[df_dados_covid['semana epidemiologica'] == semana_especifica]
por_semana = df_semana_especifica.groupby('estado')[['novos_confirmados']].sum()
por_semana
```

```python
semana_especifica = 202103
df_semana_desejada = df_dados_covid[df_dados_covid["semana epidemiologica"] == semana_especifica]
soma_dos_casos_semana = df_semana_desejada.groupby("estado")["ultimos confirmados"].sum()
soma_dos_casos_semana
```
